```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print("Features scaled successfully!")
```

The features have been scaled using `StandardScaler`. Now, we can proceed with training the Support Vector Machine (SVM) models.

```python
from sklearn.svm import SVC

# Initialize SVC with a linear kernel and C=1.0
svc_linear = SVC(kernel='linear', C=1.0, random_state=42)

# Train the model
svc_linear.fit(X_train_scaled, y_train)

# Make predictions on the scaled test data
y_pred_linear = svc_linear.predict(X_test_scaled)

# Evaluate the model
accuracy_linear = accuracy_score(y_test, y_pred_linear)
conf_matrix_linear = confusion_matrix(y_test, y_pred_linear)
class_report_linear = classification_report(y_test, y_pred_linear)

print(f"Accuracy (Linear Kernel, C=1.0): {accuracy_linear:.4f}")
print("\nConfusion Matrix (Linear Kernel, C=1.0):\n", conf_matrix_linear)
print("\nClassification Report (Linear Kernel, C=1.0):\n", class_report_linear)
```
